```python
def GetMiddleStr(target,frontStr,backStr): # obtain middle string
  frontIndex = target.index(frontStr)
  if frontIndex>=0:
    frontIndex += len(frontStr)
  backIndex = target.index(backStr)
  return target[frontIndex:backIndex]

def ToSec(date): # tarnsfor time to digital number
    datetime1 = datetime.strptime(date, '%d %b %Y %H:%M:%S') #date format in this database
    datenum = datetime1.timestamp()
    return datenum
```

# 5.Write a function which returns the percentage of emails sent on a weekend (i.e., Saturday and Sunday) as a float between 0 and 1 [3]

```python
numOfAllEmail = collection.find({'headers.Date':{'$exists': 1}}).count() # obtain Data with Date and counting
numOfSatEmail = collection.find({'headers.Date':{'$regex': "Sat,"}}).count() # obtain Data with Date(Saturday) and counting
numOfSunEmail = collection.find({'headers.Date':{'$regex': "Sun,"}}).count() # obtain Data with Date(Sunday) and counting
percentage = (numOfSatEmail+numOfSunEmail)/numOfAllEmail
print ('Total num of email sent with Date tag: ' + str(numOfAllEmail))
print ('Total num of email sent in weekend: ' + str(numOfSatEmail+numOfSunEmail))
print ('Percentage of emails sent on a weekend: ' + str(percentage))
```
